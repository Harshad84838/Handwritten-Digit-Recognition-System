```python
import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
from keras.datasets import mnist
```

```python
(x_train,y_train),(x_test,y_test)=mnist.load_data()
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_train[0]
```

```python
plt.imshow(x_train[0],cmap='binary')
```

```python
#normalisation
x_train=x_train/255
x_test=x_test/255
```

```python
x_test[0]
```

```python
model=keras.models.Sequential([
    keras.layers.Input(shape=(28,28)),
    keras.layers.Flatten(),
    keras.layers.Dense(64,activation='relu'),
    keras.layers.Dense(10,activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=10,batch_size=32,validation_split=0.1)
```

```python
test_loss,test_acc=model.evaluate(x_test,y_test)
```

```python
print(f"Test Accuracy ={test_acc*100:2f}%")
```

```python
 # Get the model's predictions for the test data
predictions = model.predict(x_test)

# Look at the prediction for the first test image
print("The model predicts this is a:", np.argmax(predictions[0]))
print("The actual label is:", y_test[0])
```

```python
# Show the image the model just guessed
plt.imshow(x_test[0], cmap='binary')
plt.title(f"Model Guessed: {np.argmax(predictions[0])}")
plt.show()
```

```python
import cv2
```

```python
image_path="/content/my_number.png.png"
img=cv2.imread(image_path,cv2.IMREAD_GRAYSCALE)
img_resized=cv2.resize(img, (28,28))

img_normalized=img_resized/255.0
img_final=np.expand_dims(img_normalized,axis=0)
plt.imshow(img_final[0],cmap='binary')


plt.show()
prediction=model.predict(img_final)
guessed_number=np.argmax(prediction)
print("The number guessed by the predictor is: ",guessed_number)
```

```python
image_path="/content/my_number2.png"
img=cv2.imread(image_path,cv2.IMREAD_GRAYSCALE)
img_resized=cv2.resize(img, (28,28))
img_inverted=cv2.bitwise_not(img_resized)

img_normalized=img_inverted/255.0
img_final=np.expand_dims(img_normalized,axis=0)
plt.imshow(img_final[0],cmap='binary')



plt.show()
prediction=model.predict(img_final)
guessed_number=np.argmax(prediction)
print("The number guessed by the predictor is: ",guessed_number)
```
